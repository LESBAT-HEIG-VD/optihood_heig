```python
def plotParetoFronts(costList1, envList1, costList2, envList2):
    plt.figure()
    plt.plot(costList1, envList1, 'o-.', label='group')
    plt.plot(costList2, envList2, 'o-.', label='indiv')
    plt.xlabel('Costs (CHF)')
    plt.ylabel('Emissions (kgCO2eq)')
    plt.title('Pareto-front-comparison')
    plt.grid(True)
    plt.legend()
    plt.savefig("ParetoFrontComparison.png",dpi=300,bbox_inches='tight')
    plt.show()        
    
plotParetoFronts(costList1, envList1, costList2, envList2)
```
